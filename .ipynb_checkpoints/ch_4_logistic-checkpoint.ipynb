## Chapter- 4: Classification - The Logistic Model

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import logit
```

```python
df = pd.read_excel('Default.xlsx', usecols=range(1,5))
df['defaultBi'] = df.default.factorize()[0]           #changes yes or no to 0 or 1
df['studentBi'] = df.student.factorize()[0]
df.head()
```

We want to predict the probablity of default by predictors of balance, income, and being a student.
Before running the regression, the yes or no must be changed to numberical values: 0 or 1.

```python
from IPython.display import Image #Allows us to import images in different formats
Image('balance.png')
```

Table 4-1: The book presents (R) estimated coefficients of the logistic regression model that predicts the probability of default using the balance (above.)
Using Python, we get the same results by running regression by statsmodel (below.)

```python
result = logit("defaultBi ~ balance", df).fit()
print(result.summary())
```

```python
result1 = logit("defaultBi ~ income", df).fit()
print(result1.summary())
```

Table 4-2: Book's (R) estimated coefficients of the logistic regression
model that predicts the probability of default using student predictor.

```python
Image('student.png')
```

```python
result2 = logit("defaultBi ~ studentBi", df).fit()
print(result2.summary())
```

```python
print (result2.conf_int()) #Interval
```

```python
result.predict(df)[0] #probability of the first observation
```

```python
Image('estimate.png')
```

```python
#p(X) = e0+1X1++pXp / 1 + e0+1X1++pXp
def prob (balance, intercept, coef):
    e = 2.71828
    return e**(intercept + coef*balance) / (1 + (e**(intercept + coef*balance)))
prob(1073.549164, -10.6513, 0.0055)
```

```python
result.pred_table() #Equivalent confusion in sikit
```

```python
results = logit("defaultBi ~ balance + income + studentBi", df).fit()
print(results.summary2())
```
